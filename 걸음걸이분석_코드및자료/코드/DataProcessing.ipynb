```python
import pandas as pd
import os
import shutil

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error: Creating directory. ' + directory)

# 데이터 디렉토리        
directory_root = './93-human-gait-database/93_human_gait_db/preprocessed_full'
# 디렉토리가 없으면 생성 있으면 넘김.
createFolder(directory_root + '/subjects_s1')
createFolder(directory_root + '/subjects_s2')


for root, dirs, files in os.walk(directory_root + '/subjects'):
    for fname in files:
        full_fname = os.path.join(root,fname)
        
        #print(full_fname)
        fname = fname.strip()
        #print(fname)
        
        if 's1' in fname:
            shutil.copy(full_fname, directory_root + '/subjects_s1/' + fname)
            
        if 's2' in fname:
            shutil.copy(full_fname, directory_root + '/subjects_s2/' + fname)
        
        
```

```python
# 두개가 아니라 하나면 지우는 소스(s1)
check_value = 0
b_fname = ""
b_sub = ""
for root, dirs, files in os.walk(directory_root + '/subjects_s1'):
    for fname in files:
        fname = fname.strip()
        fname_set = fname.split('-')
        
        if b_sub != fname_set[0]:
        
            if check_value == 1:
                #print(root)
                print(b_sub)
                print(b_fname)
                f_fname = os.path.join(root,b_fname)
                shutil.move(f_fname, directory_root + '/subjects_s1_trash/' + b_fname)
            
            check_value = 0
            
        b_sub = fname_set[0]
        b_fname = fname
        check_value += 1
```

```python
# 두개가 아니라 하나면 지우는 소스(s2)
check_value = 0
b_fname = ""
b_sub = ""
for root, dirs, files in os.walk(directory_root + '/subjects_s2'):
    for fname in files:
        fname = fname.strip()
        fname_set = fname.split('-')
        
        if b_sub != fname_set[0]:
        
            if check_value == 1:
                #print(root)
                print(b_sub)
                print(b_fname)
                f_fname = os.path.join(root,b_fname)
                shutil.move(f_fname, directory_root + '/subjects_s2_trash/' + b_fname)
            
            check_value = 0
            
        b_sub = fname_set[0]
        b_fname = fname
        check_value += 1
```

```python
import pandas as pd
import os
import shutil

# lw와 rp를 합치는 소스

# 데이터 디렉토리        
directory_root = './93-human-gait-database/93_human_gait_db/preprocessed_full'
cnt = 0
# 전처리한 데이터를 저장할 디렉터리
save_dir_root = './human-gait-dataset/'
for root, dirs, files in os.walk(directory_root + '/subjects_s1'):
    print(root)
    for i in range(0,len(files), 2):
        print(files[i], files[i+1])
        
        # 저장할 이름
        save_file_name = files[i].split('-')[0]
        save_file_name += "_s1.csv"
        print(save_file_name)
        
        # 합칠 파일들을 불러서 저장
        # csv_file1은 왼쪽 허리에서 측정한 값
        # csv_file2은 오른쪽 허벅지에서 측정한 값
        csv_file1 = pd.read_csv(root+"/"+files[i])
        csv_file2 = pd.read_csv(root+"/"+files[i+1])
        
        # 전체를 읽어온 후 gyroTimestamp로 정렬
        csv_file1 = csv_file1.sort_values(by=['gyroTimestamp_sinceReboot.s.'])
        csv_file2 = csv_file2.sort_values(by=['gyroTimestamp_sinceReboot.s.'])
        
        # 정렬 후 사용할 컬럼 선택
        # gyroTimestamp_sinceReboot.s.
        # gyroRotationX.rad.s.
        # gyroRotationY.rad.s.
        # gyroRotationZ.rad.s.
        csv_file1 = csv_file1[["gyroRotationX.rad.s.", "gyroRotationY.rad.s.", "gyroRotationZ.rad.s."]]
        csv_file2 = csv_file2[["gyroRotationX.rad.s.", "gyroRotationY.rad.s.", "gyroRotationZ.rad.s."]]
        
        # 절대값으로 변경
        #csv_file1 = abs(csv_file1)
        #csv_file2 = abs(csv_file2)
        
        # 평균값 계산
        #print(csv_file1.mean())
        #print(csv_file2.mean())
        
        # csv_file1와 csv_file2를 병합
        csv_data = pd.concat([csv_file1, csv_file2], axis=1, sort=False)
        
        # 병합후 칼럼 이름 재정의
        csv_data.columns = ['lw_x', 'lw_y', 'lw_z', 'rp_x', 'rp_y', 'rp_z']
        print(csv_data.head())
        
        # 결측값이 있는 행 제거
        csv_data = csv_data.dropna(axis=0)
        
        # 전처리한 데이터를 csv파일로 변환하여 저장
        csv_data.to_csv(save_dir_root+save_file_name, mode='w')
    

        
```
